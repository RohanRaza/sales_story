```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("month_sales.csv")
data
```

```python
plt.subplots(4,1, figsize = (6,8))
```

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(4,1, figsize = (10,12))
axes = (ax1,ax2,ax3,ax4)
for i in axes:
    i.plot(data["Months"],data["sales"], color = "#4000ff",alpha = 0.2 )
    i.set_xticks([])
    i.set_yticks([])
    
    border = ["left","right","top","bottom"]
    for j in border:
        i.spines[j].set_visible(False)
        
ax1.plot(data["Months"][0:3],data["sales"][0:3], linewidth = 2.5)
ax2.plot(data["Months"][2:6],data["sales"][2:6], linewidth = 2.5)
ax3.plot(data["Months"][5:9],data["sales"][5:9], linewidth = 2.5)
ax4.plot(data["Months"][8:12],data["sales"][8:12], linewidth = 2.5)


ax1.text(-0.1 , 20000 , "14236.9", size = 11, weight = "bold", color = "#490E6F" )
ax1.text(2.3 , 60000 , "55691.01", size = 11, weight = "bold", color = "#490E6F")

ax2.text(2.3 , 60000 , "55691.01", size = 11, weight = "bold", color = "#490E6F")
ax2.text(5 , 40000 , "34595.13", size = 11, weight = "bold", color = "#490E6F")

ax3.text(5 , 40000 , "34595.13", size = 11, weight = "bold", color = "#490E6F")
ax3.text(8 , 85000 , "81777.35", size = 11, weight = "bold", color = "#490E6F")
  
ax4.text(8 , 85000 , "81777.35", size = 11, weight = "bold", color = "#490E6F")
ax4.text(12 , 75000 , "69545.62", size = 11, weight = "bold", color = "#490E6F")

ax1.text(2.4,10000,"Jan-Mar", rotation = 60, size = 13, weight = "bold", color = "#99bbff")
ax2.text(3.4,7600,"Apr-Jun", rotation = -10, size = 13, weight = "bold", color = "#99bbff")
ax3.text(7.4,40000,"Jul-Sep", rotation = 60, size = 13, weight = "bold", color = "#99bbff")
ax4.text(10.5,40000,"Oct-Dec", rotation = 60, size = 13, weight = "bold", color = "#99bbff")

ax1.text(3.5, 100000, "Quarterly Performance Review of Market", size = 22, weight = "bold", color = "#003cb3")
ax1.text(3.5, 85000, "Total Sales in Each Quarter", size = 16, weight = "bold", color = "#003cb3")
```
